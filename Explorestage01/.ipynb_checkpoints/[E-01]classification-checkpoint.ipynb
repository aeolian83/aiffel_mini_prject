```python
# modules import( )
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
```

### (6) Evaluate the model(  )

How should we interpret the predictions of the trained models on the test data? What would be a good metric to evaluate the performance of the models? Please select an appropriate evaluation metric from sklearn.metrics. Also, explain your choice.

       ?      ? sklearn.metrics        .    .

- Precision: The proportion of instances classified as True by the model out of the actual True instances is called the true positive rate.  
    ( True     True  .)
- Recall: The proportion of instances predicted as True by the model out of the actual True instances is called the recall or true positive rate.  
    ( True    True   .)
- The F1 score is a evaluation metric composed of precision and recall, and it is a well-performing metric in imbalanced data. The reason is that the F1 score is the harmonic mean of precision and recall, designed to be sensitive to the lower of the two metrics.  
    (F1 score Precision Recall ,     .  F1 score Precision Recall         . )

- Due to the importance of recognition performance in digit recognition, I have determined that comparing the overall performance metric, the F1 score, is more advantageous than precision or recall. I created a DataFrame to compare the F1 scores and visualized the results with a graph.  

    (      precision recall   f1-score    . 
    f1-score     DateFrame   . )

### (7)-1 Initial conclusion on wine classification( 1 )
- The results are quite questionable.(  .)
- The recall value of a specific model is significantly low, even though the data is not severely imbalanced in one direction. In contrast, the accuracy remains almost the same across the other models, with a noticeable decrease in accuracy for that particular model.  
    ( recall   .      ,           .)
- After careful consideration, I have come to the conclusion that data normalization is necessary.(      .)
-    1    . 
    reference_Mr.lee [ ](https://github.com/seuyon0101/aiffel/blob/master/%5BE-01%5Dclassification.ipynb)
- Mr. Lee concluded that min-max scaling would be more effective than standardization based on a comparison of Gaussian distributions. Therefore, I performed data normalization using min-max scaling.  
    ( gaussian   standardization minmax scailing     minmax scaling   .)

```python
# Data normalization   by 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

ex_wine_data_sc = ex_wine.data
scaler.fit(ex_wine_data_sc)
scaled_wine_data = scaler.transform(ex_wine_data_sc)

scaled_wine_data.shape
```

```python
# Training data generation  
x_wine_train_sc, x_wine_test_sc, y_wine_train_sc, y_wine_test_sc = train_test_split(scaled_wine_data, 
                                                                                    ex_wine_label, 
                                                                                    test_size=0.3, 
                                                                                    random_state=11)
```

```python
# Training and prediction for each model(    )

# Decision Tree
ex_wine_decision_tree_sc = DecisionTreeClassifier(random_state=11)
ex_wine_decision_tree_sc.fit(x_wine_train_sc, y_wine_train_sc)
y_wine_dt_pred_sc = ex_wine_decision_tree_sc.predict(x_wine_test_sc)

ex_wine_decision_tree_score_sc = classification_report(y_wine_test_sc, y_wine_dt_pred_sc, output_dict=True)
df_wine_decision_tree_sc = pd.DataFrame(ex_wine_decision_tree_score_sc).transpose()

# Random Forest
ex_wine_random_forest_sc = RandomForestClassifier(random_state=11)
ex_wine_random_forest_sc.fit(x_wine_train_sc, y_wine_train_sc)
y_wine_rf_pred_sc = ex_wine_random_forest_sc.predict(x_wine_test_sc)

ex_wine_random_forest_score_sc = classification_report(y_wine_test_sc, y_wine_rf_pred_sc, output_dict=True)
df_wine_random_forest_sc = pd.DataFrame(ex_wine_random_forest_score_sc).transpose()

# Support Vector Machine
ex_wine_svm_model_sc = svm.SVC()
ex_wine_svm_model_sc.fit(x_wine_train_sc, y_wine_train_sc)
y_wine_svm_pred_sc = ex_wine_svm_model_sc.predict(x_wine_test_sc)

ex_wine_svm_model_score_sc = classification_report(y_wine_test_sc, y_wine_svm_pred_sc, output_dict=True)
df_wine_svm_model_sc = pd.DataFrame(ex_wine_svm_model_score_sc).transpose()

# Stochastic Gradient Descent Classifier(SGDClassifier)
ex_wine_sgd_model_sc = SGDClassifier()
ex_wine_sgd_model_sc.fit(x_wine_train_sc, y_wine_train_sc)
y_wine_sgd_pred_sc = ex_wine_sgd_model.predict(x_wine_test_sc)

ex_wine_sgd_model_score_sc = classification_report(y_wine_test_sc, y_wine_sgd_pred_sc, output_dict=True)
df_wine_sgd_model_sc = pd.DataFrame(ex_wine_sgd_model_score_sc).transpose()

# Logistic Regression
ex_wine_lr_model_sc = LogisticRegression()
ex_wine_lr_model_sc.fit(x_wine_train_sc, y_wine_train_sc)
y_wine_lr_pred_sc = ex_wine_lr_model.predict(x_wine_test_sc)

ex_wine_lr_model_score_sc = classification_report(y_wine_test_sc, y_wine_lr_pred_sc, output_dict=True)
df_wine_lr_model_sc = pd.DataFrame(ex_wine_lr_model_score_sc).transpose()
```

```python
# Analysis of the results( )

pr = "precision"
rc = "recall"
f1 = "f1-score"
model_columns = ['Decision Tree', 'Random Forest', 'Support Vector Machine', 'SGDClassifier', 'Logistic Regression']


df_wine_model_sc = (pd.DataFrame(data=(df_wine_decision_tree_sc[f1], 
                                      df_wine_random_forest_sc[f1], 
                                      df_wine_svm_model_sc[f1], 
                                      df_wine_sgd_model_sc[f1], 
                                      df_wine_lr_model_sc[f1]), index=model_columns).transpose())


df_wine_model_sc
```
